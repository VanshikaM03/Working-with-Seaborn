```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

**Distribution** **Plots**

```python
 sns.distplot(tips['total_bill'],kde=False,bins=30) #works on single variable #KDE-kernel density estimation
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') #takes two variables
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm') #make a jointplot of every two possible variables
```

```python
sns.rugplot(tips['total_bill']) #also works on single variable
```

```python
sns.kdeplot(tips['total_bill'])
```

**Categorical Plots**

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips) #category wise segregates the numerical input
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False,hue='sex') #y is numeric
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
 sns.catplot(x='day',y='total_bill',data=tips,kind='bar') #use instead of factorplot
```

**Matrix Plots**

```python
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc=tips.corr()
tc
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights.corr()
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=1)
```

```python
sns.clustermap(fp) #clustered together on the basis of similarity
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

**Grid**

```python
sns.pairplot(iris)
sns.PairGrid(iris)
```

```python
g=sns.PairGrid(iris) #creates empty grids to plot any type of map on it
import matplotlib.pyplot as plt
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')#scatter takes two arguments
```

**Regression Plots**

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex',aspect=0.8)
```
